```python
#set the style for seaborn
sns.set_style("whitegrid")

#Plot the histogram for numerical features
plt.figure(figsize=(12,8))
df.drop("Id" , axis=1).hist(edgecolor="black" , linewidth=1.2)
plt.suptitle("Histogram of Numerical Features")
plt.show()

#Plot Boxplot for numerical features
plt.figure(figsize=(12,8))
sns.boxplot(data=df.drop("Id" , axis=1))
plt.title("Boxplot of Numerical Features")
plt.show()

#Plots Scatter plot to explore the relationships between features
plt.figure(figsize=(12,8))
sns.pairplot(df.drop("Id" , axis=1) , hue="Species" , markers=["o" , "s" , "D"])
plt.suptitle("Pairplot of Numerical features")
plt.show()

#Barplot for distribution of species

plt.figure(figsize=(12,8))
sns.countplot(data=df.drop("Id", axis=1) , x='Species')
plt.title("Distribution of species")
plt.show()
```
